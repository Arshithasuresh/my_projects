```python
import nltk
from nltk.corpus import wordnet

# Download WordNet data if not already downloaded
nltk.download('wordnet')

# Function to find the most similar words from WordNet
def autocorrect(word, max_suggestions=3):
    synsets = wordnet.synsets(word)
    suggestions = []
    if synsets:
        for synset in synsets[:max_suggestions]:
            for lemma in synset.lemmas()[:max_suggestions]:
                suggestions.append(lemma.name())
    return suggestions if suggestions else None

# Main function
def main():
    while True:
        user_input = input("Enter any word: ")
        corrected_words = []
        for word in user_input.split():
            suggestions = autocorrect(word)
            if suggestions:
                corrected_words.append(", ".join(suggestions))
            else:
                corrected_words.append(f"No correction found for '{word}'")
        print(' '.join(corrected_words))

if __name__ == "__main__":
    main()
```
